```python
import pandas as pd
from sklearn import linear_model
```

```python
dataset=pd.read_csv(r"C:\Users\arpit\Downloads\homeprices (1).csv")
dataset
```

```python
#using pandas to create dummy variable
dummies=pd.get_dummies(dataset.town).astype(int) #it makes dummy for town from dataset (we use .astype(int) to get 0 and 1 w/o it T and F)
dummies#here data is showing true and false instead of 0 and 1 but will behave as 0 and 1
```

```python
merged= pd.concat([dataset,dummies],axis='columns') #used to merge data columnwise
merged
```

```python
final= merged.drop(['town','monroe township'],axis='columns')#used to drop columns from data
#more than one column dropped to avoid dummy variable trap
final
```

```python
x=final.drop(['price'],axis='columns')
y=final.price
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
model=reg
```

```python
model.predict(x)#predict for all x
```

```python
model.predict([[2600,0,0]])
```

```python
model.score(x,y)#calculate the score
```
