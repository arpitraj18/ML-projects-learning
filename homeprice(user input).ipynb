```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\ARPIT\Downloads\homeprices.csv")
df
```

```python
#we have a missing data to solve this we will take median data
```

```python
ndf=df.bedrooms.median()
```

```python
df.bedrooms.fillna(ndf)
#filling nan value with the median
```

```python
df.bedrooms=df.bedrooms.fillna(ndf)
df
#updatind the dataset
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
#fit is used to train model using the provided dataset
#syntax of train= reg.fit(df[['independent variabl 1','independent variabl 2']],df.target_variable)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
#price for 3000 area, 3 bedroom and 40 year old
```
