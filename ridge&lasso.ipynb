```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv(r"C:\Users\arpit\Downloads\Melbourne_housing_FULL.csv")
dataset
```

```python
dataset.nunique()
```

```python
dataset.isna().sum()# shows us no. of na in following columns
```

```python
#limiting the dataset only to useful data
#dropping irrelavant data such as date etc
useful_columns=['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
```

```python
dataset=dataset[useful_columns]
dataset
```

```python
dataset.isna().sum()
```

```python
columns_zero=['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']#these columns are selected
dataset[columns_zero]=dataset[columns_zero].fillna(0)#in all those columns all the na values are filled with 0
```

```python
#now filling building area and land size with mean values
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())#fills all the na values in landsize with mean of landsize
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
print(dataset.dtypes)
```

```python
dummies=pd.get_dummies(dataset, drop_first=True).astype(int) #does encoding and drops first
dummies
```

```python
x = dummies.drop('Price', axis=1)#axis=1 used for column, if we had to drop row then axis=0
y = dummies['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
#If random_state is set to a fixed number (like 2): The same sequence of random numbers will be generated every time the code is run, which means the same result each time
```

```python
#now using linear regression on train data
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

```python
#we are experiencing an overfitting condition here
#without random state there was severe overfitting (absurd score was given i.e.(-147719830931.68997))
#always use random state to train model with same data everytime
```

```python
#now using LASSO regression(L1)
```

```python

from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=50, max_iter=100, tol=0.1)#we can play around with alpha value to find out which gives best
lasso_reg.fit(x_train,y_train)
#alpha is the lamda value we use in our cost function formula
#The max_iter parameter specifies the maximum number of iterations the algorithm will run for optimization. 100 iterations means the algorithm will run up to 100 times to optimize the model. If the solution is found before 100 iterations, it will stop early.generally 100 iterations is enough for most problems
#The tol (tolerance) parameter specifies the stopping criterion for the optimization algorithm. The algorithm will stop when the change in the loss function between iterations is smaller than this value. A smaller tol value (e.g., 0.01) would require a more precise solution (smaller changes), whereas a larger tol allows for faster convergence with a less accurate solution.
```

```python
lasso_reg.score(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
#we see there is no overfitting
```

```python
#now using RIDGE regression (L2)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```
