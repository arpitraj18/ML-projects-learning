```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:/Users/arpit/Downloads/homeprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price)
plt.xlabel('area(sq. ft)')
plt.ylabel('price($)')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)



```

```python
x = reg.predict([[5000]])
x
```

```python
inr=85.51*x
inr
```

```python
reg.coef_
```

```python
reg.intercept_
```
